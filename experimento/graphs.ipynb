```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams['figure.figsize'] = [25, 15]
plt.rcParams['font.size'] = 30
```

```python
df = pd.read_csv('MVA-MS.csv', sep=';')
```

```python
df = df.apply(lambda x: x.astype('str').str.replace(',', '.'))
```

```python
fig, ax = plt.subplots()

x = df.index.values
form = '-4+0.0284*2*x'
y = eval(form)
D = 0.0027*2+0.002*5+0.005*2+0.0222*1+0.0284*2
D_arr = [D]*1000

knee_pos = np.where(y <= D)[-1][-1]

r1 = plt.plot(df['R'].astype('float64'), label='R x n', linewidth=5.0)
r2 = plt.plot(y, color='r', label='Assntota Dmax', linewidth=2.0)
r3 = plt.plot(D_arr, color='#EB6F18', label='Assntota D', linewidth=5.0)
r4 = plt.plot([knee_pos]*1000, df['R'].astype('float64'), color='0.75', label='Joelho', linewidth=5.0)

plt.title('Variao de R de acordo com o nmero de clientes')
plt.ylabel('R')
plt.xlabel('Nmero de clientes')
# plt.ylim([0, 55])
# plt.xlim([-1, 13])

plt.legend()
plt.tight_layout()
plt.savefig('R_n.png')
```
